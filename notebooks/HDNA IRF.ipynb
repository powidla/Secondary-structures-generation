```python
import os
import pandas as pd
from bs4 import BeautifulSoup

def extract_score(html_file):
    with open(html_file, 'r') as f:
        soup = BeautifulSoup(f, 'html.parser')

        rows = soup.find_all('tr')
        for row in rows[1:]:
            columns = row.find_all('td')
            if len(columns) >= 8:

                score = columns[7].text.strip()
                try:
                    score_float = float(score)
                    return score_float
                except ValueError:
                    print(f"Unable to convert {score} to float in file: {html_file}")
                    return None
    return None

html_folder = '/content/Hdna_vae'

scores = []

for filename in os.listdir(html_folder):
    if filename.endswith('.html'):
        html_file = os.path.join(html_folder, filename)
        score = extract_score(html_file)
        if score is not None:
            scores.append(score)

df = pd.DataFrame({'Score': scores})
```

```python
mean_score = df['Score'].mean()
std_score = df['Score'].std()
```

```python
print(f"Mean: {mean_score}")
print(f"Standard Deviation: {std_score}")
```
